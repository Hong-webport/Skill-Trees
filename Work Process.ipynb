```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# WorkProcess

A = nx.DiGraph()

# W.add_node(' ')

A.add_node('Agile')
A.add_node('SCRUM')
A.add_node('Kanban')

# W.add_node('S5')
# W.add_node('Sort')
# W.add_node('Set in Order')
# W.add_node('Shine')
# W.add_node('Standardize')
# W.add_node('Sustain')

A.add_edges_from([('Manifesto', 'Agile'),('Agile','SCRUM'),('Agile','Kanban')])
A.add_edges_from([('Kanban','1. Backlog'),('Kanban','2. Plan'),('Kanban','3. Doing'),('Kanban','4. Review'),('Kanban','5. Done')])

A.add_edges_from([('Stories','Epics'),('Epics','Initiatives'),('Initiatives','Agile') ])

A.add_edges_from([('SCRUM','Participants'),('SCRUM','Ceremonies'),('SCRUM','Artifacts')])
A.add_edges_from([('Participants','Product Owner/Manager'),('Participants','SCRUM Master/Lead'),('Participants','Members')])
A.add_edges_from([('Ceremonies','Two-Week Sprint'),('Ceremonies','Review and Retrospective'),('Ceremonies','Daily Scrum')])
A.add_edges_from([('Artifacts','Product Backlog'),('Artifacts','Sprint Backlog'),('Artifacts','Product Increment'),('Artifacts','Burndown Chart')])


A.add_edges_from([('Individuals/Interactions over Processes/Tools','Manifesto'),('Working Software over Documentation','Manifesto')
                    ,('Customer Collaboration over Negotiation','Manifesto'),('Responding to Change over Plan','Manifesto')])


# A.add_edges_from([('DevOps','1. Discover'),('DevOps','2. Plan'),('DevOps','3. Build'),('DevOps','4. Test'),('DevOps','5. Deploy'), ('DevOps','6. Operate'),('DevOps','7. Observe'),('DevOps','8. Continuous feedback')])


# pos_W = nx.nx_agraph.graphviz_layout(W, prog='dot', args="-Grankdir=LR")
# pos_W = nx.spectral_layout(W)
pos_A = nx.nx_agraph.graphviz_layout(A, prog='circo')

# pos_M = nx.spring_layout(M, k=0.30, iterations=20)
plt.figure(figsize=(20, 15))
# Draw the graph with customized features
nx.draw_networkx(A,
    pos=pos_A,  with_labels=True,
    node_size=2500,
    node_color='white',
    edge_color='skyblue',
    width=1,
    arrows=True,
    arrowstyle='->',
    font_size=8,
    font_color='red',
    alpha=0.8,
    node_shape='o')

plt.axis('off') 
plt.show()
```
