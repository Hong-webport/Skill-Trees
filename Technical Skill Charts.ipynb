```python
#python char
# Create an empty graph
G = nx.DiGraph()
# Add nodes to the graph
G.add_node('Python')

G.add_node('Data Tools')

G.add_node('Pandas')
G.add_node('SQLAlchemy')
G.add_node('Json')
G.add_node('Requests')

G.add_node('Calculations')

G.add_node('Sklearns')
G.add_node('Scipy')
G.add_node('Numpy')

G.add_node('Web')
G.add_node('App Frameworks')
G.add_node('Scraping')

G.add_node('Django')
G.add_node('Flask')
G.add_node('PySpark')

G.add_node('Scraping')

G.add_node('Selenium')

G.add_node('Beautiful Soup')
G.add_node('Splinter')
G.add_node('Browser')

G.add_node('Visualization')

G.add_node('Matplotlibs')
G.add_node('Seaborn')
G.add_node('Plotly')

G.add_node('Graphviz')
G.add_node('PyGraphviz')
G.add_node('Networkx')

G.add_node('GIS')
G.add_node('Geopy')
G.add_node('Geocoders')
# G.add_node('Nominatum')
G.add_node('Folium')

```

```python
#add edges 
# Add edges between nodes
G.add_edges_from([('Python','Data Tools'), ('Python','Calculations'), ('Python','Visualization'), ('Python','Web'),  ('Web','App Frameworks'), ('Web','Scraping') ])

G.add_edges_from([('Data Tools', 'Pandas'), ('Data Tools', 'SQLAlchemy'), ('Data Tools', 'Json'), ('Data Tools', 'Requests')])

G.add_edges_from([('Visualization', 'Matplotlibs'), ('Matplotlibs', 'Seaborn'), ('Visualization', 'Plotly')])

G.add_edges_from([('Visualization', 'Graphviz'), ('Graphviz', 'PyGraphviz'), ('Graphviz', 'Networkx')])

G.add_edges_from([('Visualization', 'GIS'), ('GIS', 'Geopy'), ('Geopy', 'Geocoders'), ('GIS', 'Folium') ])

G.add_edges_from([('Calculations', 'Sklearns'), ('Calculations', 'Scipy'), ('Calculations', 'Numpy')])

G.add_edges_from([('Scraping', 'Beautiful Soup'), ('Scraping', 'Splinter'), ('Splinter', 'Browser'), ('Scraping', 'Selenium')])

G.add_edges_from([('App Frameworks', 'Django'), ('App Frameworks', 'Flask'), ('App Frameworks', 'PySpark'), ])
```

```python


# Define the layout
# pos = nx.spring_layout(G)
pos = nx.nx_agraph.graphviz_layout(G, prog='dot',)

plt.figure(figsize=(25, 10))
# Customize the plot

# Draw the graph with customized features
nx.draw_networkx(
    G,
    pos=pos,
    with_labels=True,
    node_size=3000,
    node_color='skyblue',
    edge_color='darkgreen',
    width=1.5,
    arrows=True,
    arrowstyle='-',
    font_size=12,
    font_color='black',
    
    alpha=0.5,
    node_shape='p'
)



# # Show the plot
# plt.show()
plt.savefig('Pythons Libraries 07182023.png', dpi=300)
```
